```python
import pandas as pd
import gc

from sklearn.model_selection import train_test_split
```

```python
features = pd.read_parquet(r"C:\Users\PNC\Desktop\Coding\AI-Contest\CTR_Prediction\data\train.parquet", engine="pyarrow")

obj_col = features.columns[features.dtypes == "object"][:5]

features[obj_col] = features[obj_col].astype("float32")

train_na_cols = features.columns[features.isna().any()]

for col in train_na_cols:
    features[col].fillna(0, inplace=True)
    features[f"{col}_is_missing"] = features[col].isna().astype("float32")

del obj_col, train_na_cols

gc.collect()
```

```python
target = features.pop("clicked").to_frame()

X_train, X_val, y_train, y_val = train_test_split(
    features, target, test_size=0.2, random_state=7, stratify=target
)


DATA_PATH = r"F:/CTR_Prediction/data/"

X_train.to_parquet(DATA_PATH + "X_train.parquet", engine="pyarrow", index=False, row_group_size=100000)
X_val.to_parquet(DATA_PATH + "X_val.parquet", engine="pyarrow", index=False, row_group_size=100000)
y_train.to_parquet(DATA_PATH + "y_train.parquet", engine="pyarrow", index=False, row_group_size=100000)
y_val.to_parquet(DATA_PATH + "y_val.parquet", engine="pyarrow", index=False, row_group_size=100000)
```
